# Join  e tratamentos dos DataFrames

```python
import pandas as pd
```

```python
# Abrindo o dataset com dados de clima
clima = pd.read_csv('clima_SemEpi.csv', sep = ',')
clima.head()
```

```python
# Abrindo o dataset com os dados de status das ovitrampas
ovi = pd.read_csv('ovi_SemEpi.csv', sep = ',')
ovi.head()
```

```python
# Fazendo o join das planilhas de ovitrampas com os dados meteorológicos
ovi_clima = ovi.join(clima.set_index('SemEpi'), on='SemEpi')
```

```python
ovi_clima.head()
```

```python
# Verificando os tipos das variáveis
print (ovi_clima.dtypes)
```

```python
# procurando valores nulos
print(ovi_clima.isnull().sum())
```

```python
# procurando células vazias
print(ovi_clima.isna().sum())
```

```python
# Visualizar os dados estatísticos
ovi_clima.describe()
```

```python
# Salvando o dataset unido
ovi_clima.to_csv('ovi_clima.csv', encoding='utf-8', index=False)
```
