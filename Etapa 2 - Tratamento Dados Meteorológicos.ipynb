# CARREGAMENTO E TRATAMENTO DE DADOS

```python
import pandas as pd
```

```python
# Abrindo arquivo
arq = pd.read_csv('INMET_SE_MG_F501_BELO HORIZONTE - CERCADINHO_01-01-2017_A_31-12-2017.csv', sep = ';', decimal = ',')
```

```python
arq.head()
```

```python
# Verificar as informaçãoes do dataset
arq.info()
```

```python
# Eliminando colunas que não serão usadas
arq2 = arq.drop(columns=['HORA (UTC)','PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)','PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)', 'PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)', 'RADIACAO GLOBAL (KJ/m²)', 'TEMPERATURA DO PONTO DE ORVALHO (°C)', 'TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)', 'TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)', 'VENTO, DIREÇÃO HORARIA (gr) (° (gr))', 'VENTO, RAJADA MAXIMA (m/s)', 'VENTO, VELOCIDADE HORARIA (m/s)', 'UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)', 'UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)'])
```

```python
arq2.head()
```

```python
# Renomeando colunas
arq2.rename(columns={'DATA (YYYY-MM-DD)': 'data','PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'precip', 'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'temp', 'TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)': 'temp_min', 'TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)':'temp_max', 'UMIDADE RELATIVA DO AR, HORARIA (%)': 'umid'}, inplace = True)
```

```python
arq2.head(10)
```

```python
# Removendo valores negativos (menores que zero) do DataFrame

f_remove = arq2.loc[(arq2['precip'] < -1) | (arq2['temp_min'] < -1) | (arq2['temp_max'] < -1) | (arq2['temp'] < -1) | ((arq2['umid'] < -1))]
```

```python
arq3 = arq2.drop(f_remove.index)
```

```python
arq3.head(10)
```

```python
# Procurando valores nulos
print(arq3.isnull().sum())
```

```python
# procurando células vazias
print(arq3.isna().sum())
```

```python
# Verificando os tipos das variáveis
print (arq3.dtypes)
```

```python
# Análise estatístca do dataset para verificação de incorfomidades
arq3.describe()
```

```python
# Visualizando o shape do dataset
arq3.shape
```

```python
# Convertendo data para datatime
arq3['data'] = pd.to_datetime(arq3.data)
```

```python
# Checando os tipos das variáveis
print (arq3.dtypes)
```

```python
# Agrpudando as datas iguais pela média
arq4 = arq3.groupby('data').mean()
```

```python
arq4.head()
```

```python
# Visualizar o shape do dataset
arq4.shape
```

```python
arq4 = arq4.reset_index()
```

```python
arq4.head(10)
```

```python
# Agrupando em semanas

df = (arq4
     .set_index('data')
     .groupby([pd.Grouper(freq = 'W')]).mean()
     .reset_index())
df
```

```python
# Arredandomento dos valores
df = df.round(3)
```

```python
# Criação de um nova coluna com a identificação da semana epidemiológica
import random
semanas = list(range(1, 54, 1))
```

```python
print(len(semanas))
```

```python
# Inserindo a coluna SemEpi
df.insert(loc=5, column = 'SemEpi', value = semanas)
```

```python
df.head(10)
```

```python
# Salvando datset tratado
df.to_csv('clima_SemEpi.csv', encoding='utf-8', index=False)
```
